```python
def sigmoid(z):
    g=1/(1+np.exp(-z))
    return g
```

```python
print(f"sigmoid({0})= ",sigmoid(0))
print ("sigmoid([ -1, 0, 1, 2]) = " + str(sigmoid(np.array([-1, 0, 1, 2]))))
```

```python
def cost_function(x,w,b,y,*argv):
    m,n=x.shape
    cost = 0.0
    for i in range(m):
        z_i = np.dot(w,x[i]) + b
        f_wb_i = sigmoid(z_i)
        cost += -y[i]*np.log(f_wb_i) - (1-y[i])*np.log(1-f_wb_i)
             
    cost = cost / m
    return cost
```

```python
def gradient(x,y,w,b,*argv):
    m,n=x.shape
    dj_dw=np.zeros(w.shape)
    dj_db=0.
    for i in range(m):
        f_wbi=sigmoid(np.dot(x[i],w)+b)
        err_i=f_wbi-y[i]
        for j in range(n):
            dj_dw[j]+=err_i*x[i,j]
        dj_db+=err_i
    dj_dw/=m
    dj_db/=m
    return dj_dw,dj_db
```

```python
def compute_gradient_descent(x,y,w,b,num_itr,alpha,gradient,cost_function,lambda_):
    J_history=[]
    w_history=[]
    for i in range(num_itr):
        dj_dw,dj_db=gradient(x,y,w,b,lambda_)
        w-=alpha*dj_dw
        b-=alpha*dj_db
        if i<1000000: 
            cost =  cost_function(x, w, b,y,lambda_)
            J_history.append(cost)
        if i% math.ceil(num_itr/10000) == 0 or i == (num_itr-1):
            w_history.append(w)
            print(f"Iteration {i:4}: Cost {float(J_history[-1]):8.2f}   ")
        
    return w, b
```
